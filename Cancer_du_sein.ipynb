```python
#Extraction de la cible 'y' à partir du Dataframe.
y = BC_df_nouvo['target']
y
```

```python
#Extraction de 'X' à partir du Dataframe.
X = BC_df_nouvo.drop(['target'], axis=1)
X
```

```python
#Division du DataFrame en données d'entraînement et données de test (80% pour l'entraînement et 20% pour le test).

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Normalisation les données d'entraînement et de test avec la méthode vue en classe.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

##**Partie 3: Modélisation**

```python
# Construction, entraînement et test du 1er algorithme : K–Nearest Neighbor
%%time
from sklearn.neighbors import KNeighborsClassifier

#Construction
knn_classifier = KNeighborsClassifier()
#Entrainement
knn_classifier.fit(X_train_scaled, y_train)
```

```python
# Construction, entraînement et test du 2ème algorithme : Decision Tree Classifier

%%time
from sklearn.tree import DecisionTreeClassifier
#Construction
dt_classifier = DecisionTreeClassifier()
#Entrainement
dt_classifier.fit(X_train_scaled, y_train)
```

```python
# Construction, entraînement et test du 3ème algorithme : Random Forest Classifier
%%time
from sklearn.ensemble import RandomForestClassifier

#Construction
rf_classifier = RandomForestClassifier()
#Entrainement
rf_classifier.fit(X_train_scaled, y_train)
```

```python
#Affichage de la matrice de confusion pour les trois algorithmes que vous avez choisi.

# premier algorithme

#Etape 1 : y_pred_knn
y_pred_knn = knn_classifier.predict(X_test_scaled)

#Etape 2 : Calcul de la matrice de confusion
from sklearn.metrics import confusion_matrix
cm_knn = confusion_matrix(y_test, y_pred_knn)
print(cm_knn)


```

```python
# Deuxième algorithme

#Etape 1 : y_pred_knn
y_pred_dt = dt_classifier.predict(X_test_scaled)
#Etape 2 :Calcul de la matrice de confusion
from sklearn.metrics import confusion_matrix
cm_dt = confusion_matrix(y_test, y_pred_dt)
print(cm_dt)
```

```python
# Troisième algorithme
#Etape 1 : Test
y_pred_rf = rf_classifier.predict(X_test_scaled)
#Etape 2 Calcul de la matrice de confusion
from sklearn.metrics import confusion_matrix
cm_rf = confusion_matrix(y_test, y_pred_rf)
print(cm_rf)
```

```python
#Affichage des metriques accuracy_score, precision_score, recall_score et f1_score pour les trois algorithmes.

# premier algorithme : modèle knn

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy_score_knn = accuracy_score(y_test, y_pred_knn)
precision_score_knn = precision_score(y_test, y_pred_knn, average='macro')
recall_score_knn = recall_score(y_test, y_pred_knn, average='macro')
f1_score_knn = f1_score(y_test, y_pred_knn, average='macro')

print("accuracy knn = ",accuracy_score_knn)
print("precision knn = ",precision_score_knn)
print("recall knn = ",recall_score_knn)
print("f1 knn = ",f1_score_knn)


```

```python
# Deuxième algorithme : modèle dt.

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy_score_dt = accuracy_score(y_test, y_pred_dt)
precision_score_dt = precision_score(y_test, y_pred_dt, average='macro')
recall_score_dt = recall_score(y_test, y_pred_dt, average='macro')
f1_score_dt = f1_score(y_test, y_pred_dt, average='macro')

print("accuracy dt = ",accuracy_score_dt)
print("precision dt = ",precision_score_dt)
print("recall dt = ",recall_score_dt)
print("f1 dt = ",f1_score_dt)
```

```python
# Troisième algorithme : modèle Random Forest Classifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy_score_rf = accuracy_score(y_test, y_pred_rf)
precision_score_rf = precision_score(y_test, y_pred_rf, average='macro')
recall_score_rf = recall_score(y_test, y_pred_rf, average='macro')
f1_score_rf = f1_score(y_test, y_pred_rf, average='macro')

print("accuracy rf = ",accuracy_score_rf)
print("precision rf = ",precision_score_rf)
print("recall rf = ",recall_score_rf)
print("f1 rf = ",f1_score_rf)
```

## Partir 3 : Interprétation des Résultats

```python
#Comparaison entre trois algorithmes en utilisant la bibliothèque matplotlib pour tracer "bar chart".
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
fig.set_size_inches(10,5)
ax.set(ylim=[0.9, 1.0])

score_algo1 = [accuracy_score_knn, precision_score_knn, recall_score_knn, f1_score_knn]
score_algo2 = [accuracy_score_dt,  precision_score_dt, recall_score_dt, f1_score_dt]
score_algo3 = [accuracy_score_rf, precision_score_rf, recall_score_rf, f1_score_rf]

barWidth = 0.1  # the width of the bars

# Set position of bar on X axis
r1 = np.arange(4)
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Make the plot
plt.bar(r1, score_algo1, color='red', width=barWidth, edgecolor='white', label='algo1')
plt.bar(r2, score_algo2, color='blue', width=barWidth, edgecolor='white', label='algo2')
plt.bar(r3, score_algo3, color='orange', width=barWidth, edgecolor='white', label='algo3')

# Add xticks on the middle of the group bars
plt.xlabel('Resultats', fontweight='bold')
plt.xticks([r + barWidth for r in range(4)], ['accuracy', 'precision', 'recall', 'f1'])

# Create legend & Show graphic
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
